## environment 

前置环境要求:python3, anaconda3

### 本地环境

最好使用 [pyenv][link] 来管理多个 python 环境

> 以下 <[anaconda3-5.3.1][link]> 均为示例版本, 实际安装可根据自己需要

```bash

brew update
brew install pyenv
pyenv install anaconda3-5.3.1
cd <你的 jupyter 工作目录>
pyenv local anaconda3-5.3.1

```

此时文件夹下会生成一个 `.python-version` 的文件, 文件内容为 `anaconda3-5.3.1`, 这时候, 如果你使用的 terminal 和 sh 有相应的主题, 可以看见自己当前目录应用的 python 版本

<img src="https://cdn.jsdelivr.net/gh/chiicy/images/44494bdfd80a2c596d068aca3b21e1241590950866.jpg" width="750"/>

### 远程环境

想搭建 [jupyter lab][link] 远程服务的话没有必要通过 [pyenv][link] 来控制 python 版本, 只要在服务器安装一个版本就够了.

```bash

wget https://repo.anaconda.com/archive/Anaconda3-2020.02-Linux-x86_64.sh #下载相应的脚本(Anaconda3-2020.02-Linux-x86_64.sh可更换)
bash ~/Anaconda3-2020.02-Linux-x86_64.sh # 执行安装脚本

```

## 安装

推荐使用 conda 安装, conda 是 anaconda 提供的命令行, 如果没有, 可能是没有执行 source 或者 anaconda 安装脚本没能将 PATH 写入到 .bashrc 或者 .zshrc 中.

### 安装 [jupyter lab][link]

```bash

conda install -c conda-forge jupyterlab

```

安装完可以用 

```bash

jupyter lab 

```

启动 jupyter lab. 启动成功后即可继续下面步骤


### 安装 [nteract][link]

nteract 是和 jupyter 类似的软件,有自己的 app, 如果已经使用了 jupyter 的话也有 jupyter lab 插件, 颜值更高, 不过功能相应就少了.

```bash

pip install nteract_on_jupyter

```

安装 nteract 之后, 启动方式就变成

```bash

jupyter nteract

```

### 安装要用到的各种插件

jupyter lab 拥有丰富的插件系统, 使用起来和一般的编辑器比如 vscode 已经很接近了, 我正在使用的插件有

- [@jupyterlab/toc][link] 该插件可以生成 markdown 的目录
